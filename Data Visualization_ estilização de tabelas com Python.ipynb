```python
# Importar a biblioteca
import pandas as pd
```

```python
# Ler um arquivo CSV
df = pd.read_csv('/content/relatorio_vendas.csv')
df
```

```python
# Verificar se os dados estão nuloa
df.info()
```

```python
# Transformar data_pedido para datetime
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
```

```python
# Verificar se os dados estão nuloa
df.info()
```

Missãi 1: Time de vendas

```python
df_cliente_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```python
df_cliente_vendas.columns = ['Clientes', 'Vendas']
df_cliente_vendas['Ranque'] = df_cliente_vendas.index + 1
df_cliente_vendas.set_index('Ranque', inplace=True)
df_cliente_vendas
```

Objeto Styler permite aplicar estilos formatados a um Dataframe

* Cores do fundo
* Cores do texto
* Bordas
* Alinhamento

```python
s = df_cliente_vendas.style
s
```

```python
s.format({'Vendas': 'R$ {:,.2f}'})
s
```

# Destacando elementos

Missão 2: Time comercial

```python
df_venda_lucro = df.groupby(['tipo_produto'])[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'tipo produto'
df_venda_lucro
```

```python
estilo_produto = df_venda_lucro.style
estilo_produto
```

```python
estilo_produto.format('R$ {:,.2f}').highlight_max(color='lightgreen').highlight_min(color ='#F16165')
```

```python
estilo_produto.format('R$ {:,.2f}').background_gradient(cmap='Greens')
```

```python
cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: center; text-transform: capitalize;'
}
estilo_produto.set_table_styles([cabecalho], overwrite=False)
```

# Visualizando distribuições

```python
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N° de pedidos']
df_regiao
```

```python
df_regiao.index.name = 'Regiao'
```

```python
import numpy as np
```

```python
porcentagem = df_regiao['N° de pedidos'].to_numpy()
porcentagem = 100 * porcentagem / porcentagem.sum()
porcentagem
```

```python
df_regiao['Porcentagem regiao'] = porcentagem
df_regiao
```

```python
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: right; background-color: white'
}

celulas = {
    'selector': 'td',
    'props': 'background-color: white;'
}
```

```python
estilo_regiao.set_table_styles([cabecalho,celulas])
```

```python
estilo_regiao.format({'Porcentagem regiao': '{:,.2f} %'})\
  .bar(subset='Porcentagem regiao', vmin=0, vmax=100.0, color='#9CD33B')
```

# Lidando com muitos dados

Missão 4: Time de logística

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

```python
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses = df_meses.reset_index(drop=True)
df_meses
```

```python
vendas_mensais = df_meses.pivot_table(index = 'departamento', columns = 'meses', values = 'quantidade',
                                      aggfunc = 'sum', sort = False)
vendas_mensais
```

```python
estilo_mensal = vendas_mensais.style
estilo_mensal
```

```python
estilo_mensal.set_sticky(axis = 'index')
```

```python
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: normal; font-family: Arial'
}

tabelas = {
    'selector': 'td, th',
    'props': 'text-align: left;'
}

index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}

estilo_mensal.set_table_styles([colunas, tabelas, index], overwrite=False)
```

# Relatório de performance

Missão 5: Relatório de performance

```python
df_cliente = df.pivot_table(index = 'segmento_cliente', columns = 'modo_envio', values = 'vendas',
                            aggfunc = 'sum')
df_cliente
```

```python
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total'] = df_cliente.sum()
df_cliente
```

```python
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
tabela = {
    'selector': 'td,th:not(.index_name)',
    'props': 'font-weight: normal; font-family: Arial; text-align: center; background-color: white;'
}

index = {
    'selector': '.index_name',
    'props': 'font-weight: normal; text-align: rigth; font-style: italic; color: #696969'
}
compra_cliente.set_table_styles([tabela,index])
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }],
    'B2B': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }]
}, overwrite = False, axis=1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite=False, axis=0)
```

```python
cores_coluna = pd.DataFrame(['false', 'true', 'false'], index= df_cliente['Total'].index, columns=['Total'])
compra_cliente.set_td_classes(cores_coluna)
```

Desafio final

```python
compra_cliente.set_table_styles({
    'Total': [{'selector': '.true', 'props': 'background-color: #D8D8D8;'}]
}, overwrite=False, axis=1)

cores_linha = pd.DataFrame([['false', 'false', 'true ', 'false', 'false']],
                            columns=df_cliente.columns,
                            index=['Total'])

compra_cliente.set_td_classes(cores_linha)
```
