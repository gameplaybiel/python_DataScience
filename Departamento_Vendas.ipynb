Uma loja de departamentos deseja controlar as vendas de três dos seus principais departamentos e compreender como estão dispostas as vendas por mês de cada um deles. Carol, como uma profissional que atua nessa empresa, está responsável por obter esses dados e apresentar a composição de cada departamento frente às vendas da loja.

Inicialmente, ela requisitou os dados da base da empresa e construiu o DataFrame abaixo com os dados extraídos:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vendas = pd.DataFrame({'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
                               'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
                       'Automotivo': [45141.6, 27921.26, 96376.11, 49361.8, 43791.83, 49255.57,
                                      79187.47, 55769.71, 86045.24, 97583.11, 158102.04, 74656.09],
                       'Jardinagem e paisagismo': [71416.24, 20368.49, 42999.74, 47464.38, 45804.96, 35571.88,
                                                   41961.53, 99862.62, 96883.25, 70243.34, 101816.55, 99248.93],
                       'Materiais de construção': [17960.17, 24528.58, 34607.92, 17887.47, 46345.31, 53734.45,
                                                   47680.44, 50691.40, 95668.84, 58946.42, 100493.08, 117211.83]})
vendas.set_index("Mês", inplace=True)
vendas
```

```python
fig, ax = plt.subplots(figsize=(14,7))
azul, verde, cinza = "#3a86ff","#06d6a0","#8d99ae"
vendas.plot(kind="area", stacked=True, color = [azul, verde, cinza], ax = ax, fontsize = 12)

ax.set_title('Vendas mensais da loja de departamentos em 2022 (em reais)', fontsize = 16)
ax.set_xlabel('Mês', fontsize=14)
ax.set_ylabel('Valor (em R$)', fontsize=14)
ax.tick_params(axis='both', which='both', length=0)
ax.yaxis.grid(linestyle='--', linewidth=0.3)
ax.set_frame_on(False)
plt.show()
```
