```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregado = pd.read_sql(query, engine)
```

```python
empregado
```

```python
empregado.to_sql('empregado', con=engine, index=False)
```

```python
pd.read_sql_table('empregado', engine)
```

```python
pd.read_sql_table('empregado', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```
